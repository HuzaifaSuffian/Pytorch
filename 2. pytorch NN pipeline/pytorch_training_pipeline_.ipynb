```python
import numpy as np
import pandas as pd
import torch

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(['id','Unnamed: 32'],axis=1,inplace=True)
```

```python
# Train test split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2)
```

```python
# Standardize the input
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_test_scaled
```

```python
# Label Encoding
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)
```

```python
y_train_encoded
```

# Converting to pytorch tensor

```python
X_train_tensor = torch.from_numpy(X_train_scaled)
X_test_tensor = torch.from_numpy(X_test_scaled)
y_train_tensor = torch.from_numpy(y_train_encoded)
y_test_tensor = torch.from_numpy(y_test_encoded)
```

# Defining the model

```python
class myNN:

  def __init__(self,X):
    self.weights = torch.rand(X.shape[1],1, dtype=torch.float64, requires_grad=True)
    self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)

  def forward(self,X):
    z = torch.matmul(X,self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  def loss_func(self,y_pred,y_train):
      epsilon = 1e-7
      y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

      # Calculate loss
      loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
      return loss
```

```python
# Define initial weights and biases
# Init imp parameters
# forward prop
# loss calc
# update param
```

```python
learning_rate = 0.1
epochs = 100
```

```python
model = myNN(X_train_tensor)

for epoch in range(epochs):
  # forward prop
  y_pred = model.forward(X_train_tensor)

  # calculate loss
  loss = model.loss_func(y_pred, y_train_tensor)

  # backward prop
  loss.backward()

  # update param
  with torch.no_grad():
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad

  # grad zero
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  # print the loss
  print(f'Epochs: {epochs+1} Loss: {loss.item()}')
```

```python
# Model Evaluate
with torch.no_grad():
  y_pred = model.forward(X_test_tensor)
  y_pred = (y_pred > 0.9).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
